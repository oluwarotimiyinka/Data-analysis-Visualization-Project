```python
# Drop columns not needed

dataset1.drop(['reporting_entity', 'flaring_emissions_MtCO2','own_fuel_use_emissions_MtCO2', 'fugitive_methane_emissions_MtCO2e', 'fugitive_methane_emissions_MtCH4', 'source',],
            axis=1, inplace=True)
```

```python
# Creating Table with plotly express library
# Import create_table figure factory

from plotly.figure_factory import create_table 
  
colorscale = [[0, '#4d004c'], [.5, '#f2e5ff'], [1, '#ffffff']] 
table = create_table(dataset1.head(25), colorscale=colorscale) 
py.iplot(table)
```

```python
px.bar(dataset1.head(70), x = 'year', 
	y = 'total_emissions_MtCO2e',color = 'total_emissions_MtCO2e')
```

```python
# Plot Histograms for the commodity('Oil & NGL', 'Natural Gas')
# Set the style of seaborn
sns.set(style="whitegrid")
```

```python
dataset1 = dataset1.replace([np.inf, -np.inf], np.nan)
plt.figure(figsize=(20, 4))
sns.lineplot(data=dataset1, x='year', y='total_emissions_MtCO2e', hue='commodity', marker='o')
plt.title('CO2 Emissions for commodities over the years')
plt.xlabel('Year')
plt.ylabel('total_emissions_MtCO2e')
plt.legend(title='Commodities')
plt.show()
```

```python
# Create single line graph
dataset1 = dataset1.replace([np.inf, -np.inf], np.nan)
plt.figure(figsize=(20, 4))
sns.lineplot(data=dataset1.head(20), x='year', y='product_emissions_MtCO2', hue='parent_type', marker='o')
plt.title('Product Emissions for commodities over the years')
plt.xlabel('Year')
plt.ylabel('product_emissions_MtCO2')
plt.legend(title='parent_type')
plt.show()
```
